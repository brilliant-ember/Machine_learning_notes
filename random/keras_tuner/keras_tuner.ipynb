````python
def create_model(hp=None, name="model"):
    '''Creates a model and if hyperparameter tuner is passed it will use it
    Args:
        hp: keras_tuner.tuners tuner
        name: str, the name of the model
    Returns:
        tf.keras.model
    '''
    ## Hyperparameters we will try to tune
    num_dense_hidden_layers = 1
    num_units = 8
    dropout_rate = 0.1
    learning_rate = 0.01
    
    if hp:
        # hp.Choice goes over the list of choices given in the 'values' list
        num_dense_hidden_layers = hp.Choice('num_hidden_dense_layers', values = [1, 2, 3])
        num_units = hp.Choice("num_units", values=[8, 16, 32])
        #hp.Float searches for optimal value in the range of min_value and max_value
        dropout_rate = hp.Float('dropout_rate', min_value=0.1, max_value=0.5)
        learning_rate = hp.Float('learning_rate', min_value=0.0001, max_value=0.001)
        
    ins = Input(shape=(28, 28, 1))
    # normalization layer
    x = tf.keras.layers.Lambda(lambda i: i/255.0, name='normalization_layer')(ins)
    x = tf.keras.layers.Flatten()(x)
    
    for _ in range(0, num_dense_hidden_layers):
        x = Dense(num_units, activation='relu')(x)
        x = tf.keras.layers.BatchNormalization()(x)
        x = Dropout(dropout_rate)(x)
    
    x = Dense(10, activation='softmax', name='output')(x) # softmax changes the weight logits to probability distribution
    model = Model(inputs=ins, outputs=x, name=name)
    # we use sparse_categorical_crossentropy because we didn't onehot encode the input data
    # if we onehot encoded we would have used the typical categorical_crossentropy
    model.compile(loss='sparse_categorical_crossentropy', metrics=['accuracy'],
                  optimizer='adam')
    return model
    
    
````

## Now we create the custom tunning class
We want to find the perfect batch_size hyperparam

There was an old api that used to create a customTuner and inhereit from `kt.tuners`, however, it is depricated and now we will use the new api which `HyperModel` class instead. That's the main syntax difference between the old and new api.

#### the old way ( we will not use it, just here for reference)
The old api used something like this


``` python
class CustomTuner(kt.tuners.BayesianOptimization):
    def run_trial(self, trail, *args, **kwargs):
        kwargs['batch_size'] = trail.hyperparameters.Int('batch_size', 32, 128, step=32)
        super(CustomTuner, self).run_trial(trail, *args, **kwargs)
        
tuner = CustomTuner(
    create_model,
    objective="val_accuracy",
    max_trials=20,
    directory='logs',
    project_name='fashion_mnist',
    overwrite=True, 
)
# note that batch_size will not show in the summary, but it will show in the search
tuner.search_space_summary()

```

For the old api we can use one of the prebuilt tuners, but since we want to also find the optimial `batch_size` parameter
we have to do it inside a custom tuner and override the `run_trial` method. This means that the batch size result will not be shown in the summary, but it will be show in the resutls for the parameter search

----

#### The new way
 the new api (which we'll use) we need to create a class that inherits from `kt.HyperModel` and we need to override both the build and the fit functions. The build function is the same as the create_model function defined above
 
 https://keras.io/guides/keras_tuner/getting_started/
 
 https://github.com/keras-team/governance/blob/master/rfcs/20210920-tune-end-to-end-ml-workflows-in-keras-tuner.md#custom-training-loop



````python
# The new way as outlined in the keras tutorial uses kt.Hypermodel class
class MyHyperModel(kt.HyperModel):
    def build(self, hp):
        model = create_model(hp, name='tuned_model')
        return model
    def fit(self, hp, model, *args, **kwargs):
        batch_size_hp = hp.Int('batch_size', 32, 128, step=32)
        return model.fit(*args, batch_size=batch_size_hp, **kwargs)
    
    
````

````python
tuner = kt.BayesianOptimization(
    MyHyperModel(),
     objective="val_accuracy",
    max_trials=10,
    executions_per_trial=3,
    directory=logs_dir,
    project_name='fashion_mnist',
    overwrite=True, 
    
)

tuner.search_space_summary()
````

````python
tb_dir = os.path.join(logs_dir, "tb_logs")

tuner.search(x_train,y_train, validation_data=(x_test, y_test), epochs=1,
             callbacks=[tf.keras.callbacks.TensorBoard(tb_dir)] )
````
